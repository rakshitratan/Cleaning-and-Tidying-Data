# loc Function
Conditional selections with boolean arrays using data.
loc[<selection>] is the most common method that is used with Pandas DataFrames.
With boolean indexing or logical selection,
you pass an array or Series of True/False values to the 
.loc indexer to select the rows where
your Series has True values.

```python
df.head(10)
```

# This function works mostly as a character matching:
df.loc[2] # Finds all the values realted to '2'

```python
df.loc[[0,3,6]]
```

# iloc
Difference: loc gets rows (or columns) with particular labels from the index. iloc gets rows (or columns) at particular positions in the index (so it only takes integers). ix usually tries to behave like loc but falls back to behaving like iloc
if a label is not present in the index.

```python
df.iloc[2]
```

# Melting the data from wide range to long form:

We can clearly see the data is distributed on a wide scale:
    
So we will try to make it long using the melt function.


# Method_1


```python
pd.melt(df, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
       var_name='Year',
       value_name='Internet users (per 100 people)')
```

# Method_2


```python
df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
       var_name='Year',
       value_name='Internet users (per 100 people)')
```

# Clearly the data is now more clear and looks tidy

```python
df_long = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
       var_name='Year',
       value_name='Internet users (per 100 people)')
```
