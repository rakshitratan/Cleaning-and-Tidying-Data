# Visualising Missing Values:

# Dealing with missing values

Using different methods to find details about our dataset we know how many NA/missing/0 values we have. 

```python
df.isna().sum()
```

1) Dropping the NaN values

Using "Dropna" function drops all the missing values in rows by default.

```python
drop_NaN = df.dropna()
```

```python
drop_NaN.isna().sum()
```

```python
df.shape
```

```python
drop_NaN.shape
```

```python
drop_NaN
```

Dropping 'NaN' values in columns using the 'dropna' function: 

```python
drop_nan = df.dropna(axis=1)
```

```python
drop_nan.shape
```

2) Using 'Fillna' function

Filllig the NaN values with '0' using 'fillna'" function:

```python
fill_NaN = df.fillna(0)
fill_NaN.count()
```
